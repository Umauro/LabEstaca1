### 3.- Simulación

Una parte neurálgica de la estadística consiste en utilizar métodos computacionales para poder recrear,
en un ambiente de variables controladas, sucesos que en la vida real son dificiles de observar, medir o
replicar, a este tipo de aproximación se le conoce como simulación computacional.

#### 3.1 Piedra, papel o tijera
El popular juego de piedra, papel o tijera ha llegado a tal punto que se ha organizado un torneo nacional y su mejor amigo quiere participar. Su amigo ha investigado exhaustivamente en güiquipedia.es y descubrió que hay tres estrategias para jugar y le ha pedido a usted, que haga un análisis mediante simulaciones del desempeño de éstas.

##### Estrategias
<ul>
    <li>**A**: Elegir aleatoreamente mi próxima jugada.</li>
    <li>**B**: Al ganar mantengo mi elección hasta perder y el cambio es aleatorio.</li>
    <li>**C**: Mantengo mi elección ganadora, pero al perder el cambio es cíclico. Por ejemplo: si mi ciclo es piedra, tijera y luego papel; si pierdo con tijera elijo papel, si pierdo con papel elijo piedra y así sucesivamente.</li>
</ul>

Al empatar en B y C asuma que su amigo mantendrá su elección, pero su oponente hará un cambio como si hubiera perdido y que al primer lanzamiento ninguno de los dos contrincantes tiene preferencia por alguna opción. 

Por alguna mística razón, los oponentes de su amigo siempre usarán el ciclo: piedra, papel, tijera. Mientras que su amigo usará el ciclo tijera, papel, piedra.

Su labor es simular todos los encuentros posibles usando todas las estrategias para darle el mejor consejo a su amigo. Para esto realice 1000 simulaciones por encuentro y determine el porcentaje de victoria aproximado en cada caso.

#### 3.2 Método de MonteCarlo

El método de MonteCarlo es un método estadístico numérico basado en la simulación de variables aleatorias, con el fin de poder aproximar soluciones a expresiones matemáticas complejas o costosas de evaluar. [Ejemplo de utilización para obtener Pi](http://mathfaculty.fullerton.edu/mathews/n2003/montecarlopimod.html).

Utilizando este método usted debe calcular una aproximación del área que ocupa la campana de gauss centrada en 0 y comprendida entre $x \in [-4,4]$. La función que define a dicha campana es: 
$$
f(x) = \frac{e^{-\frac{x^2}{2}}}{\sqrt{2\pi}}
$$
